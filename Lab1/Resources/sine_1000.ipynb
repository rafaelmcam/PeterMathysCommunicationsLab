# 1000 Hz Sine #
<p>Generation and display of a sinusoidal waveform in Python.</p>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
fsz = (7,5)  # figure size
fsz2 = (fsz[0],fsz[1]/2.0)  # half high figure
```

```python
# initial parameters
Fs = 8000   # sampling rate
fm = 1000   # frequency of sinusoid
tlen = 1.0  # length in seconds
```

```python
# generate time axis
tt = np.arange(np.round(tlen*Fs))/float(Fs)
tt = np.linspace()
# generate sine
xt = np.sin(2*np.pi*fm*tt)
```

```python
np.linspace(0, tlen, int(np.floor(Fs / tlen)), endpoint = False)
```

```python
np.linspace(0, 0.78234, int(np.round(Fs * 0.78234)), endpoint = True)
```

```python
np.arange(np.round(0.78234*Fs))/float(Fs)
```

```python
np.arange(np.round(tlen*Fs))/float(Fs)
```

usar formato sugerido do arange para não ter problemas futuros, se é Fs temos round(Fs * duration) amostras

```python
# print the first 12 values of x(t)
print('xt = {}'.format(xt[:12]))
```

```python
plt.figure(1, figsize=fsz)
plt.plot(tt[:24], xt[:24])
plt.grid()
# plt.savefig('sine_1000a.eps')
```

```python
# create a labeled graph
plt.figure(2, figsize=fsz)
plt.plot(tt[:24], xt[:24], '-b')
plt.plot(tt[:24], xt[:24], 'or', label='xt values')
plt.ylabel('$x(t)$')
plt.xlabel('t [sec]')
strt2 = 'Sinusoidal Waveform $x(t)$'
strt2 = strt2 + ', $f_m={}$ Hz, $F_s={}$ Hz'.format(fm, Fs)
plt.title(strt2)
plt.legend()
plt.grid()
# plt.savefig('sine_1000b.eps')
```
